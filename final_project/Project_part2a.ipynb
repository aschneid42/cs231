```
### Let's start with VGG, because its simple to understand, we use it in CrossView, and that's the style of model I was building to do cifar10.
### Compared to the models I was building 2C2D -> 4C2D, VGG16 is a beast with 5C3D!! But it does start with much larger images with many more pixels. 
from tensorflow.keras.applications.vgg16 import VGG16 as Model

# Load model
# model = Model(weights=None, include_top=True, input_shape=(32,32,3), classes=10)
model = Model(weights='imagenet', include_top=True)
model.summary()
```

```
import pickle

# with open('/content/mydrive/imagenet_class_names.pickle', 'wb') as handle:
#     pickle.dump(imagenet_class_names, handle, protocol=pickle.HIGHEST_PROTOCOL)

with open('/content/mydrive/imagenet_class_names.pickle', 'rb') as handle:
    imagenet_class_names = pickle.load(handle)
```

```
### Visualizing Dense layer using ActivationMaximization ###
### Using the tf-keras-vis package, we have to define the layer we want to investigate and modify it to a linear layer for this test. 
def model_modifier(m):
    m.layers[-1].activation = tf.keras.activations.linear
```

```
### Put the model and model_modifier into this ActivationMaximization class to create an object we'll use later. 
### Here there is an interesting option to clone the model to modify or just modify the original. In this case we will not clone it, so that the softmax layer is removed
### Then when we ask teh model to make a prediction, instead of giving probabilities between 0-1 (which tend to emphasize the top score) we then output the scores themselves.
from tf_keras_vis.activation_maximization import ActivationMaximization
activation_maximization = ActivationMaximization(model,
                                                 model_modifier,
                                                 clone=False)
```

```
### Create a random noise image to start off the image generation process.
seed_input = tf.random.uniform((1, 224, 224, 3), 0, 255)
# seed_input = tf.random.uniform((1, 32, 32, 3), 0, 255)
```

```
pred=model(seed_input)
noise_pred=pred[0]
plt.plot(noise_pred,'.')
plt.title('Prediction values for random noise image: '+imagenet_class_names[np.argmax(noise_pred)])
plt.xlabel('Class index')
plt.ylabel('Score')
plt.show()
```

```
### Here we can use the model to build an image that looks like object#420, a banjo
%%time
from tf_keras_vis.utils.callbacks import Print
def loss(output):
    return output[:, 420]
activation = activation_maximization(loss,
                                     seed_input=seed_input,
                                     steps=512,
                                     callbacks=[Print(interval=50)])
```

```
### Here we see an image of what VGG16 thinks the perfect banjo might look like, 
### And according to the scores, this looks much more like a banjo than anything else. 
image = activation[0].astype(np.uint8)

subplot_args = { 'nrows': 2, 'ncols': 1, 'figsize': (4, 8),}
                #  'subplot_kw': {'xticks': [], 'yticks': []} }
f, ax = plt.subplots(**subplot_args)
ax[0].imshow(image)
ax[0].set_title(imagenet_class_names[420], fontsize=14)
ax[0].axis('off')
pred=model(tf.expand_dims(image,0))
print(np.argmax(pred[0]))
ax[1].plot(noise_pred,'.',color='k')
ax[1].plot(420,pred[0][420],'o',color='g')
ax[1].plot(pred[0],'.',color='r')
ax[1].set_xlabel('Class index')
ax[1].set_ylabel('Score')
plt.tight_layout()
plt.show()
```

```
### Next let's look at the last convolutional layer in the 5th Conv block of VGG, just before the 3 dense layers.
```

```
### This model modifier is a little more complicated 
layer_name = 'block5_conv3' # The target layer that is the last layer of VGG16.

def model_modifier(current_model):
    target_layer = current_model.get_layer(name=layer_name)
    new_model = tf.keras.Model(inputs=current_model.inputs,
                               outputs=target_layer.output)
    new_model.layers[-1].activation = tf.keras.activations.linear
    return new_model
```

```

from tf_keras_vis.activation_maximization import ActivationMaximization
activation_maximization = ActivationMaximization(model,
                                                 model_modifier,
                                                 clone=False)
```

```
### This layer has 512 filters, starting with #0. 
filter_number = 510
def loss(output):
    return output[..., filter_number]
```

```
%%time
from tf_keras_vis.utils.callbacks import Print

# Generate max activation
activation = activation_maximization(loss,
                                     steps=512,
                                     callbacks=[Print(interval=50)])
image = activation[0].astype(np.uint8)
```

```
# Here we can see some interesting structure, but it doesn't look strongly like any of the 1000 labels.
subplot_args = { 'nrows': 2, 'ncols': 1, 'figsize': (4, 8)}
                #  'subplot_kw': {'xticks': [], 'yticks': []} }
f, ax = plt.subplots(**subplot_args)
ax[0].imshow(image)
ax[0].set_title('filter[{:03d}]'.format(filter_number), fontsize=14)
ax[0].axis('off')
pred=model(tf.expand_dims(image,0))
top5=list(np.array(pred[0]).argsort()[-5:])
top5.reverse()
ax[1].plot(10,0,'.',color='r')
ax[1].plot(10,0,'.',color='r')
ax[1].plot(10,0,'.',color='r')
ax[1].plot(10,0,'.',color='r')
ax[1].plot(10,0,'.',color='r')
ax[1].plot(noise_pred,'.',color='k')
ax[1].plot(pred[0],'.',color='r')
ax[1].legend(['1st= '+imagenet_class_names[top5[0]].split(',',1)[0],
                '2nd= '+imagenet_class_names[top5[1]].split(',',1)[0],
                '3rd= '+imagenet_class_names[top5[2]].split(',',1)[0],
                '4th= '+imagenet_class_names[top5[3]].split(',',1)[0],
                '5th= '+imagenet_class_names[top5[4]].split(',',1)[0]])
ax[1].set_xlim([0,2000])
plt.tight_layout()
plt.show()
```
